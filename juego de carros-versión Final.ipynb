### Clase Jugadores

### Clase Juego

clase que realiza todas las acciones del juego, es decir: 
* Asignar los turos.
* Realizar el avance de cara carro.
* Revisar si ya llego alguien a la meta.
* Asignar los puestos. 
* contabilizar las victorias.
* revisar si ya finalizo el juego.

```python
class Juego: #clase juego
    
    def iniciar_carrera(self): #funcion que inicia la carrera 
        final=0 #guardian
        podio=[] #lista que me guarda el podio
        while final==0: #ciclo que se repite hasta que todos lleguen al final
            self.asignacion_turnos(podio) #asignacion de los turnos
            final=self.fin_carrera(podio) #funcion que busca si hay algun ganador
        return podio #regresa el podio

    def asignacion_turnos(self,podio): #funcion que asigna los turnos de cada jugador
        listaaux = listanombres[:] #lista auxiliar para asignar los turnos
        for i in range(num_j): #ciclo que se repite la cantidad de jugadores
            turno=random.choice(listaaux) #turno aleatorio de uno de los jugadores
            listaaux.remove(turno) #remover el jugador que tiene el turno de la lista auxiliar
            self.carrera(turno,podio) #comenzar la carrera

    def carrera(self,turno,podio): #funcion que realiza la carrera
        avance=random.randint(1,6)*100 #metros que debe avanzar el jugador
        if(lista_carril[listanombres.index(turno)]>0): # condicional que pregunta si el jugador aun no recorrio todo el carril
            lista_carril[listanombres.index(turno)]=lista_carril[listanombres.index(turno)]-avance #resta de distancia con el carril
        self.asignacion_podio(turno,podio) #asignacion del podio
        
    def asignacion_podio(self,turno,podio): #funcion que asigna el podio
        if(lista_carril[listanombres.index(turno)]<=0):  # condicional que pregunta si el jugador ya recorrio todo el carril
            repetido = turno in podio #booleano que guarda si esta o no en el podio el jugador
            if repetido==False: #condicion que pregunta si no esta en el podio
                podio.append(turno) #agregar el jugador al podio
        
    def fin_carrera(self,podio): #funcion que busca si ya todos llegaron a la meta
        if len(podio)==num_j: #condicion que pregunta si el contador de jugadores que ya llegaron a la meta es igual al numero de jugadores
            f=1 #cambiar guardian 

        if len(podio)!=num_j: #condicion que pregunta si el contador de jugadores que ya llegaron a la meta es igual al numero de jugadores
            f=0 #cambiar guardian 
        return f #regresar f
    
    def contar_victorias(self,listanombres,podio): #funcion que cuenta las  victorias
        lista_victorias[listanombres.index(podio[0])]=lista_victorias[listanombres.index(podio[0])]+1 #sumar una victoria al ganador
                
```

### Clase Autos

clase que crea los autos y los asigna de manera aleatoria y unica a cada jugador

```python
class Autos: #clase autos
    def crear_carros(self,num_j): #funcion que crea los autos
        #lista de los diferentes carros
        lista_carros=["Abarth","Alpine","Aston Martin","Audi","Bentley","BMW","Bugatti","Cadillac","Chevrolet","Cupra","Ferrari","Fiat","Ford","Jaguar","Lamborghini","Tesla","Mazda","maserati","Porsche","Volkswagen"]
        carros=self.asignar_autos(lista_carros) #asignamos un carro diferente a cada juagdor
        return carros #regresar carros
    
    def asignar_autos(self,lista_carros): # funcion que asigna un carro a cada jugador
        c=[] #lista para guardar los carros asiganados
        lista_aux=lista_carros #lista auxiliar
        for i in range(num_j): #ciclo que se repite la cantida de jugadores
            carro_asignado=random.choice(lista_aux) #tomar un carro aleatoriamente
            lista_aux.remove(carro_asignado) #evitar que dos jugadores tengan dos autos iguales
            c.append(carro_asignado) #agregar el carro asignado a la lista c
        return c #regresar C
```

### Codigo principal

codigo principal que crea los objetos y hace el llamado de los metodos de cada clase para realizar la correcta ejecución del juego y poder interactuar con el usuario.
